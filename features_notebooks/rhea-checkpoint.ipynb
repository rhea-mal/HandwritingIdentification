```python
## rhea thinning code
def read_jpeg_binary(path, cutoff_adjust=0, clean=True, crop=True, inversion=False):
    # Load image in grayscale
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    
    # Binarize the image using Otsu's method
    _, binarized_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Adjust threshold if cutoff_adjust is not 0
    if cutoff_adjust != 0:
        adjusted_thresh = _ + (_ * cutoff_adjust)
        _, binarized_img = cv2.threshold(img, adjusted_thresh, 255, cv2.THRESH_BINARY)
    
    # Invert image if inversion is True
    if inversion:
        binarized_img = cv2.bitwise_not(binarized_img)
    
    # Clean the image by filling white pixels with 7 or 8 black neighbors
    if clean:
        kernel = np.ones((3,3),np.uint8)
        binarized_img = cv2.morphologyEx(binarized_img, cv2.MORPH_CLOSE, kernel)
    
    # Crop the white space around the image
    if crop:
        coords = cv2.findNonZero(binarized_img)  # Find all non-zero points (text)
        x, y, w, h = cv2.boundingRect(coords)  # Find minimum spanning bounding box
        binarized_img = binarized_img[y:y+h, x:x+w]  # Crop the image
        
        # Optionally add a 1-pixel border around the cropped image
        binarized_img = cv2.copyMakeBorder(binarized_img, 1, 1, 1, 1, cv2.BORDER_CONSTANT, value=[255, 255, 255])
    
    # Convert image to binary format (0 and 1)
    binarized_img = (binarized_img / 255).astype(np.uint8)
    
    return binarized_img


image_path = 'images/student_1/w0001_s01_plND_r01.png'
binary_image = read_jpeg_binary(image_path, inversion=True)
plt.imshow(binary_image, cmap='gray')
plt.show()
```

```python
## extract characters and lines
def rotate_image(image, angle):
    """
        Rotates the given image by the input angle in the counter-clockwise direction
        Parameters
        ----------
            image : ndim np.array
                image to be rotated
            angle : float
                angle of rotation as degrees.
        Returns
        -------
            rotated image as np.array
    """
    # create an tuple that contains height/2, width/2
    image_center = tuple(np.array(image.shape[1::-1]) / 2) 
    # rot_mat 2x3 rotation mattrix
    rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
    # apply the affine transformation to the image
    # size of the output image image.shape[1::-1]
    result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
    return result

def read_image(img_path):
    image = cv2.imread(img_path)

    scale_percent = 18  # percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    rescaled_img = cv2.resize(image, dim, interpolation=cv2.INTER_AREA) 
    return image, rescaled_img  

def save_img(dir_path,filename,img):
    """
        dir_path - directory path where the image will be saved
        filename - requires a valid image format
        img - image to be saved
    """
    file_path = os.path.join(dir_path,filename)
    cv2.imwrite(file_path,img)

def find_text_angle(dilated_img,org_img):
    """
        org_img - original image
        img - dilated img
    """
    lines = cv2.HoughLinesP(dilated_img,rho=1,theta=np.pi/180,threshold=30,minLineLength=5,maxLineGap=20)

    nb_lines = len(lines)
    angle = 0

    for line in lines:
        x1,y1,x2,y2 = line[0]
        angle += math.atan2((y2-y1),(x2-x1)) 

    angle/=nb_lines

    rotated = rotate_image(org_img, angle-1)
    rot_dilated = rotate_image(dilated_img,angle-1)

    return rotated, rot_dilated
 
def extract_text_lines(img,output_dir):   
    """
        img - image from which the text lines are extracted
        output_dir - directory where the extracted lines should be saved 
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.medianBlur(gray, 5)
    thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 5, 5)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (16, 2))
    dilate = cv2.dilate(thresh, kernel, iterations=2)
    rotated,rot_dilated = find_text_angle(dilate,img)

    cnts = cv2.findContours(rot_dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(cnts) == 2:
        cnts = cnts[0]
    else:
        cnts = cnts[1]

    lines_path = os.path.join(output_dir,'lines')  

    if not os.path.exists(lines_path):
        os.makedirs(lines_path)

    for line_idx, line in enumerate(cnts, start=-len(cnts)):
        x, y, w, h = cv2.boundingRect(line)
        roi = rotated[y:y + h, x:x + w]
        filename = 'line'+str(line_idx)+ '.jpg'
        save_img(lines_path,filename=filename,img=roi)
    
def extract_text_chars(img,output_dir):
    """
        img - image from which the individual chars are extracted
        output_dir - directory where the extracted lines should be saved 
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.medianBlur(gray, 7)

    thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 7, 11)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 7))

    dilate = cv2.dilate(thresh, kernel, iterations=1)

    cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(cnts) == 2:
        cnts = cnts[0]
    else:
        cnts = cnts[1]
    
    chars_path = os.path.join(output_dir,'chars')  

    if not os.path.exists(chars_path):
        os.makedirs(chars_path)
        
    for char_idx, character in enumerate(cnts, start=-len(cnts)):
        x, y, w, h = cv2.boundingRect(character)
        roi = img[y:y + h, x:x + w]
        filename = 'char'+str(char_idx)+ '.jpg'
        save_img(chars_path,filename=filename,img=roi) 

img_file_path = 'images/student_1/w0001_s01_plND_r01.png'
img_out_dir = 'extracted_characters/student_1'
image, rescaled_image = read_image(img_path=img_file_path)
extract_text_lines(rescaled_image,img_out_dir)
extract_text_chars(image,img_out_dir)

img_file_path = 'images/student_2/w0002_s01_plND_r01.png'
img_out_dir = 'extracted_characters/student_2'
image, rescaled_image = read_image(img_path=img_file_path)
extract_text_lines(rescaled_image,img_out_dir)
extract_text_chars(image,img_out_dir)
```

```python
def calculate_average_dimensions(input_dir):
    widths, heights = [], []
    for filename in os.listdir(input_dir):
        file_path = os.path.join(input_dir, filename)
        try:
            with Image.open(file_path) as img:
                width, height = img.size
                widths.append(width)
                heights.append(height)
        except IOError:
            pass  # Skip files that are not images
    average_width = np.mean(widths)
    average_height = np.mean(heights)
    return average_width, average_height

def remove_outliers_and_save(input_dir, output_dir, width_threshold, height_threshold, density_threshold=0.01):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        file_path = os.path.join(input_dir, filename)
        try:
            with Image.open(file_path) as img:
                width, height = img.size
                if width >= width_threshold and height >= height_threshold:
                    img_gray = img.convert("L")
                    img_binary = img_gray.point(lambda x: 0 if x > 128 else 1, '1')
                    non_white_pixels = img_binary.getcolors(width * height)[0][0]
                    total_pixels = width * height
                    density = non_white_pixels / total_pixels
                    if density > density_threshold:
                        img.save(os.path.join(output_dir, filename))
        except IOError:
            print(f"Skipping file (not an image or unsupported format): {filename}")

def remove_unwanted_characters(input_dir, output_dir, density_threshold=0.01):
    average_width, average_height = calculate_average_dimensions(input_dir)
    # Define thresholds as a percentage of the average dimensions to remove outliers
    # Adjust these percentages based on your dataset's characteristics
    width_threshold = average_width * 0.5  # Example: 50% of the average width
    height_threshold = average_height * 0.5  # Example: 50% of the average height
    
    remove_outliers_and_save(input_dir, output_dir, width_threshold, height_threshold, density_threshold)

# Example usage
remove_unwanted_characters(input_dir='extracted_characters/student_1/chars', output_dir='cleaned_characters/student_1')
remove_unwanted_characters(input_dir='extracted_characters/student_2/chars', output_dir='cleaned_characters/student_2')
```

```python
def calculate_fourier_descriptors(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, 0)
    
    # Threshold the image to get a binary image
    _, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
    
    # Find contours in the binary image
    contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    # Assuming the largest contour corresponds to the character
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
    else:
        return np.array([])  # Return an empty array if no contours are found
    
    # Flatten the largest contour array and create a complex array from it
    largest_contour = largest_contour.reshape(-1, 2).astype(np.float32)
    contour_complex = largest_contour[:, 0] + 1j * largest_contour[:, 1]
    
    # Compute the Fourier Transform
    fourier_result = np.fft.fft(contour_complex)
    
    # Normalize to make the descriptors scale and rotation invariant
    fourier_descriptors = np.abs(fourier_result) / np.abs(fourier_result[0])
    
    return fourier_descriptors


# Path to the chars folder
def get_fourier_for_students(chars_folder_path):
    # Initialize a container to hold the Fourier descriptors for each character
    fourier_descriptors_list = []

    # Loop through each image in the chars folder
    for filename in os.listdir(chars_folder_path):
        if filename.endswith(".jpg"):  # Changed to accept .jpg files
            image_path = os.path.join(chars_folder_path, filename)
            
            # Calculate the Fourier Descriptors for the image
            descriptors = calculate_fourier_descriptors(image_path)
            
            # Append the descriptors to the list if they are not empty
            if descriptors.size:
                fourier_descriptors_list.append(descriptors)

    return fourier_descriptors_list

# At this point, `fourier_descriptors_list` contains the Fourier descriptors for each character image

fourier_student_1 = get_fourier_for_students('cleaned_characters/student_1')    
fourier_student_2 = get_fourier_for_students('cleaned_characters/student_2')        

print(len(fourier_student_1))
print(len(fourier_student_2))
```

```python
def calculate_lbp_features(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    # Set the number of points and the radius for LBP
    points = 8
    radius = 1
    # Compute the LBP image
    lbp = local_binary_pattern(img, points, radius, method="uniform")
    # Calculate the histogram
    (hist, _) = np.histogram(lbp.ravel(),
                             bins=np.arange(0, points + 3),
                             range=(0, points + 2))
    # Normalize the histogram
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    return hist

def calculate_canny_edges(image_path):
    # Load the image
    img = cv2.imread(image_path, 0)
    # Apply Gaussian blur to smooth the image and reduce noise
    img_blurred = cv2.GaussianBlur(img, (5, 5), 0)
    # Detect edges using the Canny detector
    edges = cv2.Canny(img_blurred, 100, 200)
    # Calculate the proportion of pixels that are edges in the image
    edge_density = np.sum(edges > 0) / (edges.shape[0] * edges.shape[1])
    return edge_density

# Example usage

def get_edge_line_features(chars_folder_path):
    lbp_features_list = []
    edge_density_list = []

    for filename in os.listdir(chars_folder_path):
        if filename.endswith(".jpg"):  # Assuming .jpg format, change accordingly
            image_path = os.path.join(chars_folder_path, filename)
            
            # Calculate LBP features
            lbp_features = calculate_lbp_features(image_path)
            lbp_features_list.append(lbp_features)
            
            # Calculate Canny edge density
            edge_density = calculate_canny_edges(image_path)
            edge_density_list.append(edge_density)
    return lbp_features_list, edge_density_list

lbp_features_list_1, edge_density_list_1 = get_edge_line_features('cleaned_characters/student_1')
lbp_features_list_2, edge_density_list_2 = get_edge_line_features('cleaned_characters/student_2')

print(len(lbp_features_list_1))
print(len(lbp_features_list_2))
print(len(edge_density_list_1))
print(len(edge_density_list_2))


# At this point, `lbp_features_list` contains the LBP histograms (texture features) for each character image
# and `edge_density_list` contains the edge density (edge features) for each character image
```

```python
# Initialize dictionaries to store the features by student
fourier_features = {}
lbp_features = {}
edge_density_features = {}

# Directory where the student folders are located
base_images_dir = 'images'

# List the directories for each student
students_dirs = [os.path.join(base_images_dir, d) for d in os.listdir(base_images_dir) if os.path.isdir(os.path.join(base_images_dir, d))]

for student_dir in students_dirs:
    student_id = os.path.basename(student_dir)  # This assumes the folder name is the student ID
    fourier_features[student_id] = []
    lbp_features[student_id] = []
    edge_density_features[student_id] = []
    
    # Output directories for extracted and cleaned characters
    extracted_dir = os.path.join('extracted_characters', student_id)
    cleaned_dir = os.path.join('cleaned_characters', student_id)
    
    # Ensure output directories exist
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    if not os.path.exists(cleaned_dir):
        os.makedirs(cleaned_dir)

    # Process each image for the current student
    for img_filename in os.listdir(student_dir):
        img_path = os.path.join(student_dir, img_filename)
        
        # Process the image
        image, rescaled_image = read_image(img_path)
        extract_text_lines(rescaled_image, extracted_dir)
        extract_text_chars(image, extracted_dir)
        remove_unwanted_characters(os.path.join(extracted_dir, 'chars'), os.path.join(cleaned_dir, 'chars'))
        
        # Calculate and store features for each character image
        char_images_dir = os.path.join(cleaned_dir, 'chars')
        for char_img_filename in os.listdir(char_images_dir):
            char_img_path = os.path.join(char_images_dir, char_img_filename)
            
            fourier_desc = calculate_fourier_descriptors(char_img_path)
            lbp_hist = calculate_lbp_features(char_img_path)
            edge_density_val = calculate_canny_edges(char_img_path)
            
            if fourier_desc.size:
                fourier_features[student_id].append(fourier_desc)
            if lbp_hist.size:
                lbp_features[student_id].append(lbp_hist)
            edge_density_features[student_id].append(edge_density_val)

# At this point, `fourier_features`, `lbp_features`, and `edge_density_features` contain the respective features
# for each student's character images, organized by student ID.
```

```python
# Assuming edge_density_features is already populated
# Convert edge density features into a format suitable for clustering
# Also keep track of labels (student IDs) for coloring
features = []
labels = []
for student_id, densities in edge_density_features.items():
    for density in densities:
        features.append([density])  # K-means expects a 2D array
        labels.append(student_id)

# Convert to numpy array for clustering
features = np.array(features)

# Apply K-means clustering
# Assuming you want as many clusters as students
num_clusters = len(edge_density_features.keys())
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(features)

# Predict cluster ids
cluster_ids = kmeans.predict(features)

# Visualization
plt.figure(figsize=(10, 6))
for student_id in edge_density_features.keys():
    # Extract this student's features and their cluster ids for plotting
    student_features = features[np.array(labels) == student_id]
    student_cluster_ids = cluster_ids[np.array(labels) == student_id]
    plt.scatter(student_features[:, 0], np.zeros_like(student_features[:, 0]), label=student_id, alpha=0.6)

# Optionally, plot the cluster centers
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers, np.zeros_like(cluster_centers), color='black', marker='x', s=100, label='Centers')

plt.title("Edge Density Features Clustering by Student")
plt.xlabel("Edge Density")
plt.yticks([])  # Hide y-axis as it's not meaningful in this 1D context
plt.legend()
plt.show()
```
